# Splitting the data into X and Y

```python
# split the dataset into dependent(X) and Independent(Y) datasets
X=df.drop(['diagnosis','id'],axis=1)
Y=df['diagnosis']
```

```python
print(X)
```

# Splitting X and Y into Train and Test 

```python
# spliting the data into trainning and test dateset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# models/ Algorithms

def models(X_train,Y_train):
    
        #logistic regression
        from sklearn.linear_model import LogisticRegression
        log=LogisticRegression(random_state=0)
        log.fit(X_train,Y_train)
        
        
        #Decision Tree
        from sklearn.tree import DecisionTreeClassifier
        tree=DecisionTreeClassifier(random_state=0,criterion="entropy")
        tree.fit(X_train,Y_train)
        
        #Random Forest
        from sklearn.ensemble import RandomForestClassifier
        forest=RandomForestClassifier(random_state=0,criterion="entropy",n_estimators=10)
        forest.fit(X_train,Y_train)
        
        #knn
        from sklearn.neighbors import KNeighborsClassifier
        knn=KNeighborsClassifier(n_neighbors=5)
        knn.fit(X_train,Y_train)
             
        #SVM
        from sklearn import svm
        SVM = svm.SVC(kernel='rbf',gamma='auto',C=2)
        SVM.fit(X_train,Y_train)
        
        #NB
        from sklearn.naive_bayes import GaussianNB
        GNB = GaussianNB()
        GNB.fit(X_train,Y_train)
        
            
        print('[0]logistic regression accuracy:',log.score(X_train,Y_train))
        print('[1]Decision tree accuracy:',tree.score(X_train,Y_train))
        print('[2]Random forest accuracy:',forest.score(X_train,Y_train))
        print('[3]KNN accuracy:',knn.score(X_train,Y_train))
        print('[4]SVM accuracy:',SVM.score(X_train,Y_train))
        print('[5]NB accuracy:',GNB.score(X_train,Y_train))
        
           
        return log,tree,forest,knn,SVM,GNB;
        
```

```python
model=models(X_train,Y_train)
```

# Testing

```python
# testing the models/result

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

for i in range(len(model)):
    print("Model",i)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print('Accuracy : ',accuracy_score(Y_test,model[i].predict(X_test)))
    print('------------------------------------------------------------')
```

```python
# prediction of Random forest
pred_dt=model[2].predict(X_test)
print('Predicted -- Y -- Target values: Random forest ')
print(*pred_dt)
print('---------------------------------------------------------------------------')


print('benign            count 0     ----------- '+ str(np.count_nonzero(pred_dt == 0)))

print('Malignant         count 1     ----------- '+ str(np.count_nonzero(pred_dt == 1)))
total_prediction = np.count_nonzero(pred_dt == 0) + np.count_nonzero(pred_dt == 1)

print('Total prediction  count 0-1   ----------- '+ str(total_prediction))

print('---------------------------------------------------------------------------')

print('Actual -- Y  -- Target values: ')

print(*Y_test)
print('---------------------------------------------------------------------------')
print('benign                     count 0     ----------- '+ str(np.count_nonzero(Y_test == 0)))
print('Malignant                  count 1     ----------- '+ str(np.count_nonzero(Y_test == 1)))
total_actual_target_values = np.count_nonzero(Y_test == 0)+np.count_nonzero(Y_test == 1)
print('Total actual target value  count 0-1   ----------- '+ str(total_actual_target_values))
```
